```python
import networkx as nx
import matplotlib.pyplot as plt
import nxviz as nv
```

```python
G = nx.Graph() 
G.add_nodes_from(['Carol', 'Andre', 'Fernando','Diane','Beverly','Ed',
                  'Garth','Heather','Ike','Jane'])
```

```python
#List of Nodes
G.nodes()
```

```python
G.add_edge('Carol','Fernando')
G.add_edge('Carol','Diane')
G.add_edge('Carol','Andre')

G.add_edge('Andre','Beverly')
G.add_edge('Andre','Diane')
G.add_edge('Andre','Fernando')

G.add_edge('Beverly','Diane')
G.add_edge('Beverly','Ed')
G.add_edge('Beverly','Garth')

G.add_edge('Garth','Ed')
G.add_edge('Garth','Fernando')
G.add_edge('Garth','Diane')
G.add_edge('Garth','Heather')

G.add_edge('Heather','Ike')
G.add_edge('Ike','Jane')
G.add_edge('Fernando','Diane')
G.add_edge('Fernando','Heather')

G.add_edge('Diane','Ed')
```

```python
#Figure 1
nx.draw(G,label=True,with_labels=True)
plt.show()
```

```python
#Figure 2
matrix_plot = nv.MatrixPlot(G)
matrix_plot.draw()
plt.show()
```

```python
#Figure 3
circos_plot = nv.CircosPlot(G)
circos_plot.draw()
plt.show()
```

```python
#Figure 3
arc_plot = nv.ArcPlot(G)
arc_plot.draw()
plt.show()
```
